```python
from pathlib import Path
from sqlalchemy import create_engine
import pandas as pd
```

```python
coffee_path = Path("Coffee_Data/.csv")

df = pd.read_csv(coffee_path, encoding="UTF-8")

df.head(10)
```

```python
df_drop = df.drop('')
df_drop_2 = df_drop.drop(df_drop.index[-1])
df_index_reset = df_drop_2.reset_index()
df_index_reset.columns = df_index_reset.iloc[0]
df_drop_3 = df_index_reset.drop([0])
df_column_drop = df_drop_3.dropna(axis=1, how='all')

df_final = df_column_drop.rename(columns={df_column_drop.columns[0]: 'Country'})
df_final = df_final.rename(columns={'Jun 2024/25': '2024/25'})

df_final.head(10)
```

```python
year_columns = ['2019/20', '2020/21', '2021/22', '2022/23', '2023/24', '2024/25']

# Convert year columns to numeric (integers), forcing errors to NaN (if any)
df_final[year_columns] = df_final[year_columns].apply(pd.to_numeric, errors='coerce', downcast='integer')

# Now create a mask for rows where all year columns are 0
mask = (df_final[year_columns] == 0).all(axis=1)

# Drop those rows from the DataFrame
df_final_filtered = df_final[~mask]

# Display the first 10 rows of the filtered DataFrame
df_final_filtered
```

```python
df_final_filtered.to_csv('.csv', index=False)
```
